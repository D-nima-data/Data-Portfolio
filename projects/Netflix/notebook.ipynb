The next qustion to answer is regarding the duration of the movies in the 90s. We can represent this data using a simple histogram. The code below provides us with a histogram with the duration of the movies released in the 90s. We can see the median duration of movies is around 100 minutes. 

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
movies_90s = netflix_df[(netflix_df['type'] == 'Movie') & (netflix_df['release_year'] >= 1990)&(netflix_df['release_year']<2000)]
all_duration = movies_90s["duration"]
print("The mean of movie durations is :", int(all_duration.mean()))
print("The median of movie durations is :", int(all_duration.median()))
#help(plt.hist)
plt.hist(all_duration, bins = 10, color = "red", histtype='step')
#plt.hist
plt.title('Movie duration of 90s movies on Netflix')
plt.xlabel('Duration')
plt.ylabel('Number of movies')
```

The final required stasistic is the number of short action movies which was described as movies with under 90 minutes of runtime. 
Using the code below we can see there were a total of 7 short movies in our data set. 

```python
#action_movies_90s = movies_90s[movies_90s['genre'] == 'Action'][["duration"]]
action_movies_90s = movies_90s[movies_90s['genre'] == 'Action'][['duration']]
#print(action_movies_90s)
movie_duration = action_movies_90s['duration']
short_movie_count = 0 
for film in movie_duration:
    if film < 90:
        short_movie_count += 1 
print("There are ",short_movie_count,"short movies in the action genre")
```
